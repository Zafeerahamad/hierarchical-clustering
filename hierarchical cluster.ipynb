```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage,dendrogram
from sklearn.datasets import make_blobs

```

```python
X, Y = make_blobs(n_samples=250, n_features=7, centers=5, cluster_std=1.0, random_state=47)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
pd.DataFrame(X)
```

```python
z=linkage(X,method="ward")
dendrogram(z)
plt.show()
```

```python
z=linkage(X,method="ward")
dendrogram(z,truncate_mode="lastp",p=10)
plt.xlabel("points")
plt.ylabel("Distance")
plt.show()
```

```python
hac=AgglomerativeClustering(n_clusters=5)
y_predict=hac.fit_predict(X)
hac.labels_
y_predict
```

```python
plt.scatter(X[:,0],X[:,1],c=y_predict)
plt.legend()
plt.show()
```
